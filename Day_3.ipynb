```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "/ign.csv"

data = pd.read_csv(path)
data.head()

```

```python
path1='/Pokemon.csv'
data1=pd.read_csv(path1)
data1.head()
data1.columns
```

```python
#remove the hash columns
P = data1.dropna(axis=1)
data1.columns
P
```

```python
plt.figure(2,figsize=(15,6))
sns.displot(data=P,x='Attack',kind='hist')
```

```python
col = ['HP',	'Attack',	'Defense'	,'Sp. Atk',	'Sp. Def',	'Speed']
plt.figure(4,figsize=(15,10))
sns.boxplot(data=data1[col])
```

```python
plt.figure(5)
plt.subplot(2,2,1)
sns.histplot(data=P,x='Attack')

plt.subplot(2,2,2)
sns.histplot(data=P,x='HP')

plt.subplot(2,2,3)
sns.histplot(data=P,x='Speed')

plt.subplot(2,2,4)
sns.histplot(data=P,x='Defense')

```

```python
# display the count of type 1 pokemons
plt.figure(6,figsize=(12,5))
ax=sns.countplot(data=P,x='Type 1')
for i in (ax.containers):
  ax.bar_label(i)
```

```python
# display the count of type 2 pokemons
plt.figure(7,figsize=(12,5))
ax=sns.countplot(data=data1,x='Type 2')
for i in (ax.containers):
  ax.bar_label(i)
```

```python
#display the number of pokemoins in each genration
#show it in a visual form
plt.figure(8,figsize=(12,5))
ax=sns.countplot(data=data1,x='Generation')
for i in (ax.containers):
  ax.bar_label(i)

```

```python
#compare the type1 pokemon with respect ti attaclk capability
plt.figure(8,figsize=(17,5))
ax=sns.barplot(data=data1,x='Type 1',y='Attack')
for i in (ax.containers):
  ax.bar_label(i)

plt.xticks(rotation=90)
```

```python
#compare the type2 pokemon with respect ti speed capability
plt.figure(8,figsize=(17,5))
ax=sns.barplot(data=data1,x='Type 2',y='Speed')
for i in (ax.containers):
  ax.bar_label(i)

plt.xticks(rotation=90)
```
