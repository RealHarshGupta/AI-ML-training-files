```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
no_of_sensors = 10
np.random.seed(5)
data = 0+(100-0)*np.random.random((no_of_sensors,2))
plt.figure(1)

plt.plot(data[:,0],data[:,1],'ro')
for i in range(data.shape[0]):
  dx = data[i,0]
  dy = data[i,1]

  plt.text(dx-1,dy+2,(int(dx),int(dy)))
```

```python
dist_matrix = np.zeros((data.shape[0],data.shape[0]))
```

```python
for i in range(data.shape[0]):
  for j in range(data.shape[0]):
     dist = np.sqrt((data[i,0]-data[j,0])**2 + (data[i,1]-data[j,1])**2)
     dist_matrix[i][j]=dist

    # dist = np.sqrt((d[i,0]-d[]))
```

```python
dist_matrix.astype(int)
```

```python
neighbour_matrix = np.zeros((dist_matrix.shape[0],dist_matrix.shape[0]))
sr=60
count_zeros=0
count_nonzeros=0
for i in range(dist_matrix.shape[0]):
  for j in range(dist_matrix.shape[0]):
    if(i!=j):
      if(dist_matrix[i][j]<60):
        neighbour_matrix[i][j]=1
        count_nonzeros=count_nonzeros+1
      else:
        count_zeros=count_zeros+1


print(neighbour_matrix)
print(count_zeros,count_nonzeros)
```

```python
node =5

neighbour_list=[]
for j in range(neighbour_matrix.shape[0]):
  if(neighbour_matrix[node-1][j]!=0):
    neighbour_list.append(j)

print(neighbour_list)
```

```python
data =0+(100-0)* np.random.random((500,2))
plt.figure(1)
plt.subplot(1,2,1)
plt.plot(data[:,0],data[:,1],'ro')
for i in range(data.shape[0]):
  if(data[i,0]<=50):
    plt.subplot(1,2,2)
    plt.plot(data[i,0],data[i,1],'bo')
  else:
    plt.subplot(1,2,2)
    plt.plot(data[i,0],data[i,1],'go')
# plt.plot(d2,'go')
```
